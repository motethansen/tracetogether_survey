# Singapore Tracetogether survey data


```python
# setup pandas and read the csv file

import pandas as pd

df_import = pd.read_excel('citizens_survey.xlsx',sheet_name='Form Responses 1')
df_import.head()
#define new header titles

df_header= {'Timestamp':'Timestamp',            
            'What is your age group':'age',
            'What is your gender?':'gender',
            'Employment status?':'employment',
            'What is your education level?':'education',
            'How many times did you test positive for COVID-19 during the pandemic?':'covid',
            'If you had COVID, how did the fact that you became ill impact your job?':'job_impact',
            'How many vaccinations did you receive?':'vaccinations',
            'How would you rank the government\'s public support during Covid (1 to 5 = very satisfied) [Finance (Support)]':'rank_financial_support',
            'How would you rank the government\'s public support during Covid (1 to 5 = very satisfied) [medical(vaccines)]':'rank_medical_vaccines',
            'How would you rank the government\'s public support during Covid (1  to 5 = very satisfied) [Public safety (social distancing)]':'rank_social_distance',
            'How effective did you find the lockdown in Singapore? (rank 1 Least to 5 Most)':'lockdown_effectiveness',
            'How would you rate the usability of the Tracetogether app?':'UX_Tracetogether',
            'How frequently did you use the TraceTogether app or Token to check into venues in Singapore?':'TT_frequency',
            'Did the TraceTogether app help you to maintain social distancing during the pandemic? ':'TT_help_socialdistance',
            'Did the TraceTogether app help reduce the spread of COVID-19?':'TT_reduce_spread',
            'Did the government provide enough information about the TraceTogether app and its benefits?':'gov_info_benefit',
            'Did the TraceTogether app enhance your sense of participation and engagement in the fight against COVID-19?':'sense_participation',
            'Did the TraceTogether app help you to feel more connected with other citizens during the pandemic?':'connected_community',
            'Did the TraceTogether app help you to feel more connected with the government during the pandemic?':'connected_gov',
            'Do you think the TraceTogether app contributed to a safer environment for citizens during the pandemic?':'TT_safer_environment',
            'Do you think the government should invest in more smart city technologies like the TraceTogether app in the future?':'more_gov_invest_sc_tech',
            'Should the government involve citizens more in developing and implementing smart city technologies in the future?':'gov_involve_citizens',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Citizens\' participation]':'factor_citizens_participation',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Public-private partnership]':'factor_PPP',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Improved lifestyle]':'factor_improved_livestyle',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Better healthcare]':'factor_better_healthcare',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Improved public transport]':'factor_improved_public_transport',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Better education opportunities]':'factor_better_education',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [More job availability]':'factor_more_jobs',
            'What factors motivate you to participate in smart city projects? [Ranking form 1 Least - 5 Most] [Other]':'factor_other',
            'If you selected Other, can you describe here?':'factor_other_txt',
            'Some cities have direct reporting or claim service for citizens. Other have social media channels where citizens can share opinions participation to be offered in your city?':'suggest_citizens_participation_txt',
            'What are the most effective ways to involve citizens in developing smart city projects? [rank 1 Least - 5 Most] [Survey]':'involvement_survey',
            'What are the most effective ways to involve citizens in developing smart city projects? [rank 1 Least - 5 Most] [City living labs]':'involvement_city_labs',
            'What are the most effective ways to involve citizens in developing smart city projects? [rank 1 Least - 5 Most] [Interviews]':'involvement_interviews',
            'What are the most effective ways to involve citizens in developing smart city projects? [rank 1 Least - 5 Most] [feedback and report solutions]':'involvement_feedback',
            'What are the most effective ways to involve citizens in developing smart city projects? [rank 1 Least - 5 Most] [inclusion with incentives]':'involvement_incentives',
            'What are the most effective ways to involve citizens in developing smart city projects? [rank 1 Least - 5 Most] [others]':'involvement_other',
            'How important is the government to involve citizens in developing smart city projects? [1 Least - 5 Most]':'importance_of_citizens_involve',
            'How would you rate the effectiveness of the government in involving citizens in developing smart city projects? [1 Least - 5 Most]':'effectiveness_citizens_involvement',
            'What factors influence citizens\' willingness to participate in smart city projects? [enter] ':'citizens_willingness_to_participate_txt',
            'How are you communicating or giving feedback to the local government?':'communication_channel_to_gov',
            'How satisfied are you with the responses from the local government, from your feedback or inquiries? [1 Least - 5 Most]':'response_satisfaction_from_gov',
            'What recommendations do you have for improving citizens\' participation in smart city projects in Singapore? [enter] ':'citizens_recommendations'
}


from typing import Iterable

def get_all_values(d):
    if isinstance(d, dict):
        for v in d.values():
            yield from get_all_values(v)
    elif isinstance(d, Iterable) and not isinstance(d, str): # or list, set, ... only
        for v in d:
            yield from get_all_values(v)
    else:
        yield d 
        
nheaderlist = []

nheaderlist=list(get_all_values(df_header))
print(nheaderlist)
df_import.columns= nheaderlist
#show all columns:
pd.set_option('display.max_columns', None)
df_import.head()
```

```python
# fix data for job impact

job_impact= {
    '(1) lost job':1,
    '(2) Reduced work':2,
    '(3) More work':3,
    '(4) No impact':4
}
# fix for ranks
rank_gov_support= {
    '1 Not at all satisfied':1,
    '1 Not applicable':0,
    '2 Not satisfied':2,
    2:2,
    3:3,
    4:4,
    'Very satisfied':5,
    '5 Very satisfied':5
}
# fix for involvement
effective_citizens_involvement = {
    '1 Least':1,
    '2':2,
    '3':3,
    '4':4,
    '5 Most':5
}

factors = {
    '5 Most':5,
    4:4,
    3:3,
    2:2,
    '1 Least':1
}

df_import["job_impact"]=df_import["job_impact"].map(job_impact)
df_import["rank_financial_support"]=df_import["rank_financial_support"].map(rank_gov_support)
df_import["rank_medical_vaccines"]=df_import["rank_medical_vaccines"].map(rank_gov_support)
df_import["rank_social_distance"]=df_import["rank_social_distance"].map(rank_gov_support)
df_import["factor_citizens_participation"]=df_import["factor_citizens_participation"].map(factors)
df_import["factor_PPP"]=df_import["factor_PPP"].map(factors)
df_import["factor_improved_livestyle"]=df_import["factor_improved_livestyle"].map(factors)
df_import["factor_better_healthcare"]=df_import["factor_better_healthcare"].map(factors)
df_import["factor_improved_public_transport"]=df_import["factor_improved_public_transport"].map(factors)
df_import["factor_better_education"]=df_import["factor_better_education"].map(factors)
df_import["factor_more_jobs"]=df_import["factor_more_jobs"].map(factors)
df_import["factor_other"]=df_import["factor_other"].map(factors)
df_import["involvement_survey"]=df_import["involvement_survey"].map(factors)
df_import["involvement_city_labs"]=df_import["involvement_city_labs"].map(factors)
df_import["involvement_interviews"]=df_import["involvement_interviews"].map(factors)
df_import["involvement_feedback"]=df_import["involvement_feedback"].map(factors)
df_import["involvement_incentives"]=df_import["involvement_incentives"].map(factors)
df_import["involvement_other"]=df_import["involvement_other"].map(factors)

df_import
```

```python
#wordcloud
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# Start with one Text column:
text_df = df_import[['citizens_willingness_to_participate_txt']]
#text_df
text=""

for key, value in text_df['citizens_willingness_to_participate_txt'].items():
    #print(key, value)
    #print("key: ", key)
    #print("value: ", value)
    text = text+" "+str(value)+" "

print(text)    

# Create and generate a word cloud image:
wordcloud = WordCloud(width = 300, height = 300,max_font_size=50, max_words=100, background_color="white", colormap = "ocean").generate(text)

# Display the generated image:
plt.figure(figsize=[15,10])
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file("wordcloud.png")
```
